```python
#Task 1 — Build the Pipeline
```

```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
data = {
    "id": list(range(1,21)),
    "title": [f"Movie {i}" for i in range(1,21)],
    "popularity": [i*5 for i in range(1,21)],
    "vote_average": [6.5 + i*0.1 for i in range(1,21)],
    "vote_count": [50 + i*10 for i in range(1,21)],
    "release_date": ["2020-01-01"]*20,
    "genre_ids": ["[28,12]","[35]","[18]","[28,53]","[16]"]*4
}

df = pd.DataFrame(data)
df.head()
```

```python
engine = create_engine("sqlite:///:memory:")
```

```python
df.to_sql("movies", engine, if_exists="replace", index=False)

print("Data saved to DB")
```

```python
#Task 2 — Perform EDA
```

```python
df = pd.read_sql("SELECT * FROM movies", engine)
df.head()
```

```python
df.describe()
```

```python
import ast

df["genre_ids"] = df["genre_ids"].apply(
    lambda x: ast.literal_eval(x) if isinstance(x, str) else x
)

genre_df = df.explode("genre_ids")

genre_df["genre_ids"].value_counts()
```
